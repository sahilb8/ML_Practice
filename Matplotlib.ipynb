```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(5,100,100)
y = np.linspace(10,1000, 100)

plt.plot(x,y)
```

```python
plt.plot(x,y)

#x, y labels and titles
plt.xlabel('current')
plt.ylabel('voltage')
plt.title('ohms LAW')

#define the range if labels of the axis
#Arguments: plt.axis(xmin,xmax,ymin,ymax)
plt.xlim([20,80])
plt.ylim([200,800])
plt.show()

```

```python
x = np.linspace(0,10,20)
y = x*2

#color blue and line type '+'
plt.plot(x,y,'b+')

#x, y labels and titles
plt.xlabel('current')
plt.ylabel('voltage')
plt.title('ohms LAW')

plt.show()
```

```python
#plot different types of lines within same plot

x = np.linspace(3,10,6)
y = np.linspace(10,20,6)

plt.plot(x,y,'r-',x**2,y,'b+',x+8,y,'g^')
plt.show()
```

```python
#create subgraphs

x = np.linspace(1,10,100)
y = np.log(x)

#initialize a new figure explicitly
plt.figure(1)

#creating a subplot with 1 row and 2 columns

#create the first subplot in figure 1
plt.subplot(121) #equivalent to plt.subplot(1,2,1)
plt.title('y=log(x)')
plt.plot(x,y)

#create the second subplot in figure 1
plt.subplot(122)
plt.title('y=log(x)**2')
plt.plot(x,y**2)

plt.show()


```

```python
x  = np.linspace(1,10,100)

plt.figure(1)

plt.subplot(221)
plt.title('linear')
plt.plot(x,x)

plt.subplot(222)
plt.title('exponential')
plt.plot(x,np.exp(x))

plt.figure(2)

plt.subplot(221)
plt.title('cubic')
plt.plot(x,x**3)

plt.subplot(222)
plt.title('log')
plt.plot(x,np.log(x))

plt.show()
```

```python
#graph with 3 rows and 4 columns
x  = np.linspace(1,10,100)

plt.figure(1)

plt.subplot(341)
plt.title('linear')
plt.plot(x,x)

plt.subplot(342)
plt.title('exponential')
plt.plot(x,np.exp(x))

plt.subplot(343)
plt.title('cubic')
plt.plot(x,x**3)

plt.subplot(344)
plt.title('log')
plt.plot(x,np.log(x))



plt.show()
```
