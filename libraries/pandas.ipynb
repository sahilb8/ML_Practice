# Pandas Dataframe

```python
temp = np.random.randint(low = 20, high = 100, size=[20,])
name = np.random.choice(['sahil','vinol','jo','chobs'], 20)
random = np.random.choice([1,2,3,4,5],20)
```

```python
a = list(zip(temp,name,random))
```

```python
#created a dataframe using list
df = pd.DataFrame(data=a, columns=['temp','name','rand'])
```

```python
df
```

```python
type(df)
```

```python
#to sort the values wrt to the index
df = df.sort_index(axis=0, ascending=False)
```

```python
df
```

```python
#to sort the values wrt columns
df.sort_values('rand', ascending=False)
```

```python
#to drop a column, be default axis=0, so change is to column(axis=1)
df.drop(['name'],axis=1)
```

```python
#here iloc will give row 2 to 7 and name column as its 1st col
df.iloc[2:8,1]
```

```python
#in loc we have to five the actual index and not starting from 0(normal), and can also specify from which column
#you need data
df.loc[[22,21],'rand':'name']
```

```python
df.loc[(df.rand>=5) | (df.rand==1)]
```

```python
df1 = pd.DataFrame({
    'city': ['delhi','panaji','mumbai','chennai','mangalore'],
    'temp': [34,32,45,12,22]
})
```

```python
df1
```

```python
df2 = pd.DataFrame({
    'city': ['panaji','mumbai','chennai'],
    'humidity': [77,88,55]
})
```

```python
df2
```

```python
#merge can be used to merge on a key and diif merges possible like left,right
pd.merge(df1,df2,on='city')
```

```python
pd.merge(df1,df2,on='city',how='left')
```
