```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns

#set a seaborn style of your taste
sns.set_style('whitegrid')
```

```python
df = pd.read_csv('/home/sahil/Desktop/UdemyMLcourseKaggleMaster/Datavisual/global_sales_data/market_fact.csv')
```

```python
df.head()
```

```python
#simple density plot
sns.distplot(df['Shipping_Cost'])
plt.show()
#this return probability on y-axis, as we can see the shipping cost of most packages are small.
```

```python
#rug plot gives us where the density is very high
sns.distplot(df['Shipping_Cost'][:200],rug=True)
plt.show()
```

```python
#to remove histogram
sns.distplot(df['Shipping_Cost'],hist=False)
plt.show()
```

```python
#subplots

plt.subplot(2,2,1)
plt.title('sales')
sns.distplot(df['Sales'])

plt.subplot(2,2,2)
plt.title('profit')
sns.distplot(df['Profit'])

plt.subplot(2,2,3)
plt.title('Order Quantity')
sns.distplot(df['Order_Quantity'])

plt.subplot(2,2,4)
plt.title('Shipping cost')
sns.distplot(df['Shipping_Cost'])

plt.show()
```

```python
#boxplot
#this boxplot will be spread over x axis
sns.boxplot(df['Order_Quantity'])
plt.title('order quantity')
plt.show()
```

```python
#to spread it over the y axis 
sns.boxplot(y=df['Order_Quantity'])
plt.title('order quantity')
plt.show()
```

```python
#Bivariate distributions are basically 2 univariate distributions plotted on x and y axis respectively.
#It shows the relationship between the 2 variables.

sns.jointplot('Sales', 'Profit', df)
plt.show()
```

```python
#to remove the outlier points
df = df[(df.Profit<20000) & (df.Sales<10000)]

sns.jointplot('Sales', 'Profit', df)
plt.show()
```
